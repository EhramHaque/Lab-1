# Task One

```python
import pandas as pd

# Read the dataset into a Pandas DataFrame
data = pd.read_csv('data.csv')
df = pd.DataFrame(data)

# Group the DataFrame by StockCode
grouped_df = df.groupby('StockCode')

# Get a list of all the unique Descriptions for each StockCode
unique_descriptions = grouped_df['Description'].unique()

# Identify products that have more than one unique description
products_with_multiple_descriptions = []
for stockcode, descriptions in unique_descriptions.items():
    if len(descriptions) > 1:
        products_with_multiple_descriptions.append(stockcode)

# Print the list of products that have more than one unique description
print(products_with_multiple_descriptions)
```

# Task 2

```python
import pandas as pd

data = pd.read_csv('data.csv')
df = pd.DataFrame(data)

# Convert 'InvoiceDate' column to datetime dtype
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Filter records for April 2021 transactions
april_2021_transactions = df[(df['InvoiceDate'].dt.year == 2021) & (df['InvoiceDate'].dt.month == 4)]

# Filter records for December 2010 transactions
december_2010_transactions = df[(df['InvoiceDate'].dt.year == 2010) & (df['InvoiceDate'].dt.month == 12)]

# Printing the results
print("Transactions occurred in April 2021:")
print(april_2021_transactions)
print("\nTransactions occurred in December 2010:")
print(december_2010_transactions)
```

# Task Three

```python
import pandas as pd

data = pd.read_csv('data.csv')
df = pd.DataFrame(data)

# Update UnitPrice for specific StockCode
stock_code = '85123A'
new_price = 5.0

# Filter rows with the specific StockCode and update UnitPrice
df.loc[df['StockCode'] == stock_code, 'UnitPrice'] = new_price

# Printing the updated DataFrame
print("Updated DataFrame:")
print(df)
```

# Task Five

```python
import pandas as pd

data = pd.read_csv('data.csv')
df = pd.DataFrame(data)

# Add a new row with intentionally incorrect data (e.g., negative Quantity)
new_row = pd.DataFrame({
    'StockCode': ['99999'],  # Assuming 99999 as a new StockCode
    'Description': ['Product F'],  # Description for the new product
    'Quantity': [-3],  # Negative Quantity
    'UnitPrice': [5.99],  # Unit Price for the new product
    'InvoiceDate': ['2023-04-15'],  # Date of the transaction
    'CustomerID': [54321],  # Customer ID
    'Country': ['Canada']  # Country of the transaction
})

# Concatenate the new row to the DataFrame
df = pd.concat([df, new_row], ignore_index=True)

# Printing the updated DataFrame
print("Updated DataFrame:")
print(df)
```
